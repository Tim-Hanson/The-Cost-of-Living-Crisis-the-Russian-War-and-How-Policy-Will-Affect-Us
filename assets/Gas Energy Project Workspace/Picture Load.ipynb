```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from glob import glob
from dotenv import load_dotenv
```

```python
Picture_Load = os.getenv('Picture_Load')
```

```python
# Glob provides a list of files from the file location

Images = glob(Picture_Load)

# Create a figure with 3 columns and 6 rows for displaying the images (3x6 grid)
fig, axes = plt.subplots(nrows=6, ncols=3, figsize=(9, 12))  # 6 rows, 3 columns
axes = axes.flatten()  # Flatten the axes for easy iteration

# Loop through the images and display them in the grid
for ax, image in zip(axes, Images):
    img = mpimg.imread(image)  # Read the image
    
    # Show the cropped image
    ax.imshow(img)
    ax.axis('off')  # Turn off the axis labels

# Add title to the entire figure
plt.suptitle("Average Energy Price as a Portion of Salary (Pence/kWH/£'000)", fontsize=16, fontweight='bold', y=1.02)

# Adjust layout for minimal whitespace between images
plt.tight_layout(pad=0.2)  # Tight layout with minimal padding between images
plt.subplots_adjust(wspace=0, hspace=0)  # Remove any space between subplots
plt.show()
```
